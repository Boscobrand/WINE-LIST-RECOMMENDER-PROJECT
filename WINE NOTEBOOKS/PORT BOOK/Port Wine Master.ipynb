```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
port_df=pd.read_csv('../scraped_data/wine_file_port_popular.csv')
```

```python
port_df
```

```python
port_df.info()
```

```python
port_df.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
port_df['wine_category']="Port"
```

```python
port_df["location"]= port_df["location"].str.replace("\n·\n", "_", case = False) 
```

```python
port_df[['country','region']] = port_df.location.apply(lambda x: pd.Series(str(x).split("_")))   
```

```python
port_df.head()
```

```python
port_df['vintage']=""
```

```python
port_df['vintage'] = port_df['wine'].map(lambda val: val[-4:])
```

```python
port_df.head(200)
```

```python
port_df['vintage'].value_counts(),
```

```python
port_wine_style_list=['Ruby','LBV','Late Bottled Vintage','White','Rosé','10','20','40','Tawny','Vintage','Crusted','Sherry','Medeira','Colheita']
```

```python
port_df['wine_style']=""
```

```python
#this is the function we will run to extract the winestyle from the wine column
def winestyle(x):
    port_df.loc[port_df['wine'].str.contains(x),'wine_style'] = x 
```

```python
for i in port_wine_style_list:
    winestyle(i)
```

```python
port_df['wine_style'] =port_df['wine_style'].replace("",'Port')
```
