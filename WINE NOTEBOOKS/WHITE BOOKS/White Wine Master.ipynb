```python
white_df=pd.concat([df_1,df_2,df_3,df_4],axis=0)
```

```python
white_df.head()
```

```python
white_df.info()
```

```python
white_df.reset_index(level=None, drop=True, inplace=True)
```

```python
white_df.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
white_df
```

```python
white_df['wine_category']="White"
```

```python
white_df["location"]= white_df["location"].str.replace("\n·\n", "_", case = False) 
```

```python
white_df[['country','region']] = white_df.location.apply(lambda x: pd.Series(str(x).split("_")))              
```

```python
white_df.head()
```

```python
white_df['vintage']=""
```

```python
def vintage (x):
    white_df.loc[white_df['wine'].str.contains(x),'vintage']= x
```

```python
#Estabish a list of vintages as strings we can loop through with our function to populate the full vintage column  
vintage_list = ['N.V.', '1985','1986','1987','1988','1989','1990', '1991', '1992','1993','1994','1995','1996','1997','1998','1999','2000','2001','2002','2003','2004','2005','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020']
```

```python
for i in vintage_list: 
    vintage (i)
```

```python
white_df['vintage'].value_counts(),
```

```python
white_wine_style_list=['Albarino','Gruner Veltliner','Muscadet','Pinot Gris','Sauvignon Blanc','Soave','Vermentino','Chardonnay','Chablis','Marsanne','Semillon','Viognier','Chenin Blanc','Gewurztraminer','Muscat','Riesling','Torrontes','Pinot Blanc','Pinot Bianco','Semillon','Fumé']
```

```python
white_df['wine_style']=""
```

```python
#this is the function we will run to extract the winestyle from the wine column
def winestyle(x):
    white_df.loc[white_df['wine'].str.contains(x),'wine_style'] = x 
```

```python
for i in white_wine_style_list:
    winestyle(i)
```

```python
white_df['wine_style'] = white_df['wine_style'].replace("",'Chardonnay')
```
