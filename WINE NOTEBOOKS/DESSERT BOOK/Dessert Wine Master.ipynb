```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dessert_df=pd.read_csv('../scraped_data/wine_file_dessert_rating.csv')
```

```python
dessert_df
```

```python
dessert_df.info()
```

```python
dessert_df.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
dessert_df['wine_category']="Dessert"
```

```python
dessert_df["location"]= dessert_df["location"].str.replace("\n·\n", "_", case = False) 
```

```python
dessert_df[['country','region']] = dessert_df.location.apply(lambda x: pd.Series(str(x).split("_")))   
```

```python
dessert_df.head()
```

```python
dessert_df['vintage']=""
```

```python
dessert_df['vintage'] = dessert_df['wine'].map(lambda val: val[-4:])
```

```python
dessert_df.head(200)
```

```python
dessert_df['vintage'].value_counts(),
```

```python
dessert_wine_style_list=['Medeira','Marsala','Sauternais','Sherry','Vin Santo','Vinsanto','Sauternes','Sauternes','Riesling','Gewürztraminer','Gewurztraminer','Icewine','Eiswein','Recioto','Sémillon','Passito','Sec','Barsac','Vignoles','Port','Moelleux']
```

```python
dessert_df['wine_style']=""
```

```python
#this is the function we will run to extract the winestyle from the wine column
def winestyle(x):
    dessert_df.loc[dessert_df['wine'].str.contains(x),'wine_style'] = x 
```

```python
for i in dessert_wine_style_list:
    winestyle(i)
```

```python
dessert_df['wine_style'] =dessert_df['wine_style'].replace("",'Dessert')
```
